```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
gas = pd.read_csv("gas_prices.csv")
gas
```

```python
values = [gas["Australia"].sum(), gas["USA"].sum(), gas["Japan"].sum(), gas["France"].sum(), gas["Italy"].sum()]
labels = ["Australia", "USA","Japan","France","Italy"]

plt.pie(values, labels=labels, autopct="%1.1f%%")
plt.title("Consumo de g√°s: Australia vs USA")
plt.savefig("country2.png")
plt.show()
```
