```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
gas = pd.read_csv("gas_prices.csv")
gas
```

```python
plt.title("Valores do combustvel em US$")

list_country = ["Australia","USA", "Canada","Japan","France"]
for country in gas:
    if country in list_country:
        plt.plot(gas["Year"],gas[country],label = [country], marker = ".")

plt.xticks(gas["Year"][::2])
plt.xlabel("Anos")
plt.ylabel("US$")
plt.legend()
plt.savefig("country.png")
plt.show()
```
